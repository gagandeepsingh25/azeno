```python
def setup_assistant(client):
    # # Upload a file with an "assistants" purpose
    # file = client.files.create(
    #     # file= open("/content/excel_code2.docx", "rb"),
    #     file= open("/content/Azenco_knowledge_base_01.docx", "rb"),
    #     purpose='assistants'
    # )

    # create a new agent
    assistant = client.beta.assistants.create(
        name="Bay Calculation",
        instructions=INSTRUCTIONS,
        tools=[
            {
                "type": "function",
                "function": {
                    "name": "calculate_bays",
                    "description": "Use this function to make Pricing calculations",
                    "parameters": {
                        "type": "object",
                        "properties": {
                            "input_string": {
                                "type": "string",
                                "description": "Use Width, Length, Product type as parameters",
                            }
                        },
                        "required": ["input_string"],
                    },
                },
            },

            {
                "type": "function",
                "function": {
                    "name": "calculate_posts",
                    "description": "Use this function for Structure type calculations",
                    "parameters": {
                        "type": "object",
                        "properties": {
                            "input_string": {
                                "type": "string",
                                "description": "Use Width and Structure type as parameters",
                            }
                        },
                        "required": ["input_string"],
                    },
                },
            },

            {
                "type": "function",
                "function": {
                    "name": "mandatory_purchase",
                    "description": "Use this function for mandatory costs calculation",
                    "parameters": {
                        "type": "object",
                        "properties": {
                            "input_string": {
                                "type": "string",
                                "description": "Use Number of bays and State name type as parameters",
                            }
                        },
                        "required": ["input_string"],
                    },
                },
            }
        ],
        model="gpt-4-1106-preview",
        # file_ids=[file.id]
    )
    return assistant
```

```python
def run_assistant(client, assistant_id, thread_id):
    # Create a new run for the given thread and assistant
    run = client.beta.threads.runs.create(
        thread_id=thread_id,
        assistant_id=assistant_id
    )
    time.sleep(1)
    # Loop until the run status is either "completed" or "requires_action"
    while run.status == "in_progress" or run.status == "queued":
        run = client.beta.threads.runs.retrieve(
            thread_id=thread_id,
            run_id=run.id
        )
        # At this point, the status is either "completed" or "requires_action"
        if run.status == "completed":
            return client.beta.threads.messages.list(
              thread_id=thread_id
            )
        if run.status == "requires_action":
              tool_call = run.required_action.submit_tool_outputs.tool_calls[0]
              tool_outputs = []
              if tool_call.function.name == "calculate_bays":

                    print("tool_call.function.arguments", tool_call.function.arguments)
                    arguments_dict = json.loads(tool_call.function.arguments)["input_string"]
                    arguments_dict = json.loads(arguments_dict)

                    arguments_dict = {key: value.replace("'", "") for key, value in arguments_dict.items()}

                    print("arguments_dict", arguments_dict)

                    processed_result = calculate_bays(arguments_dict['Width'], arguments_dict['Length'], arguments_dict['Product'])
                    print("processed_result", processed_result)
                    print(type(processed_result))
                    processed_result = str(processed_result)
                    print(type(processed_result))

                    tool_outputs.append({
                        "tool_call_id": tool_call.id,
                        "output": processed_result,
                    },
                                      )
                    # print(tool_outputs)


              elif tool_call.function.name == "calculate_posts":
                    print("tool_call.function.arguments", tool_call.function.arguments)
                    arguments_dict = json.loads(tool_call.function.arguments)["input_string"]
                    arguments_dict = json.loads(arguments_dict)

                    arguments_dict = {key: value.replace("'", "") for key, value in arguments_dict.items()}

                    print("arguments_dict", arguments_dict)

                    processed_result = calculate_posts(arguments_dict['Width'], arguments_dict['Structure'])
                    print("processed_result", processed_result)
                    print(type(processed_result))
                    processed_result = str(processed_result)
                    print(type(processed_result))

                    tool_outputs.append({
                        "tool_call_id": tool_call.id,
                        "output": processed_result,
                    },
                                      )
                    # print(tool_outputs)

              elif tool_call.function.name == "mandatory_purchase":
                    print("tool_call.function.arguments", tool_call.function.arguments)
                    arguments_dict = json.loads(tool_call.function.arguments)["input_string"]
                    arguments_dict = json.loads(arguments_dict)

                    arguments_dict = {key: value.replace("'", "") for key, value in arguments_dict.items()}

                    print("arguments_dict", arguments_dict)

                    processed_result = mandatory_purchase(arguments_dict['Bays'], arguments_dict['State'])
                    print("processed_result", processed_result)
                    print(type(processed_result))
                    processed_result = str(processed_result)
                    print(type(processed_result))

                    tool_outputs.append({
                        "tool_call_id": tool_call.id,
                        "output": processed_result,
                    },
                                      )
                    # print(tool_outputs)



              if tool_outputs:
                  run = client.beta.threads.runs.submit_tool_outputs(
                      thread_id=thread_id,
                      run_id=run.id,
                      tool_outputs=tool_outputs
                  )
```

```python
def continuous_chat(client, assistant_id):
    # Create a new thread only once
    thread = client.beta.threads.create()

    while True:
        user_input = input("User: ")

        # User sends a message
        client.beta.threads.messages.create(
            thread.id,
            role="user",
            content=str(user_input),
        )

        # Assistant responds
        messages = run_assistant(client, assistant_id, thread.id)
        # print(messages)

        if messages is not None:
            message_dict = json.loads(messages.model_dump_json())

            assistant_response = message_dict['data'][0]['content'][0]["text"]["value"]
            print(f"Assistant: {assistant_response}")
        else:
            print("No messages received from the assistant.")

if __name__ == "__main__":
    assistant = setup_assistant(client)
    continuous_chat(client, assistant.id)


#cost of installation for R-Blade, 40 x 30, NY state, Freestanding, 4 fans
```
